```python
import yfinance as yf
import pandas as pd
import os
```

```python
if os.path.exists("sp500.csv"):
    sp500 = pd.read_csv("sp500.csv", index_col=0)
else:
    sp500 = yf.Ticker("^GSPC")
    sp500 = sp500.history(period="max")
    sp500.to_csv("sp500.csv")
```

```python
sp500.index = pd.to_datetime(sp500.index)
```

```python
sp500
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500 = sp500.loc["1990-01-01":].copy()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000, min_samples_split=50, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]

    new_predictors+= [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna(subset=sp500.columns[sp500.columns != "Tomorrow"])
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1
    preds[preds <.6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_score, confusion_matrix
import matplotlib.pyplot as plt

# Naive Bayes
naive_bayes_model = GaussianNB()
naive_bayes_model.fit(train[predictors], train["Target"])

# Precision Score
naive_bayes_preds = naive_bayes_model.predict(test[predictors])
naive_bayes_precision = precision_score(test["Target"], naive_bayes_preds)
print("Precision Score (Naive Bayes):", naive_bayes_precision)

# Confusion Matrix
conf_matrix_naive_bayes = confusion_matrix(test["Target"], naive_bayes_preds)
print("Confusion Matrix (Naive Bayes):\n", conf_matrix_naive_bayes)

# Visualizing Confusion Matrix
plt.imshow(conf_matrix_naive_bayes, cmap='Blues', interpolation='nearest')
plt.title('Naive Bayes Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Predicted Probabilities for Naive Bayes
naive_bayes_probs = naive_bayes_model.predict_proba(test[predictors])[:, 1]

# Visualizing Predicted Probabilities
plt.figure(figsize=(8, 5))
plt.plot(test.index, naive_bayes_probs, label='Naive Bayes Predicted Probabilities')
plt.title('Naive Bayes Predicted Probabilities')
plt.xlabel('Date')
plt.ylabel('Probability')
plt.legend()
plt.show()

# Additional Diagnostics on Training Data for Naive Bayes
test_preds_nb = naive_bayes_model.predict(test[predictors])

# Target and Plot
test["NaiveBayes_Target"] = test_preds_nb
combined_nb = pd.concat([test["Target"], test["NaiveBayes_Target"]], axis=1)
combined_nb.columns = ["True", "Predicted"]
combined_nb.plot()
plt.title('Naive Bayes: True vs Predicted')
plt.xlabel('Date')
plt.ylabel('Target')
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import precision_score
import matplotlib.pyplot as plt

# Create GradientBoostingClassifier model
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=1)

# Train the model
gb_model.fit(train[predictors], train["Target"])

# Predictions on test data
preds_gb = gb_model.predict(test[predictors])

# Calculate precision score
precision_gb = precision_score(test["Target"], preds_gb)
print(f"Precision Score (Gradient Boosting): {precision_gb}")

# Plot true vs predicted values
test["GradientBoosting_Target"] = preds_gb
combined_gb = pd.concat([test["Target"], test["GradientBoosting_Target"]], axis=1)
combined_gb.columns = ["True", "Predicted"]
combined_gb.plot()
plt.title('Gradient Boosting: True vs Predicted')
plt.xlabel('Date')
plt.ylabel('Target')
plt.show()
```
